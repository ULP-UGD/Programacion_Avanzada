Usando Pandas y NumPy:
1. Crear un DataFrame a partir del archivo datos_metereologicos.csv
2. Realizar observaciones iniciales con pandas
3. Convertir las columnas del DataFrame en arrays de NumPy
4. Reemplazar los faltantes en los arrays por promedios respectivos
5. Analizar estadisticas sobre los arrays:
  * Temperatura promedio
  * Total de precicpitaciones
  * Máxima humedad registrada
  * La fecha más calurosa
  * La fecha más fría
6. Exportar los resultados a un archivo .csv

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ULP-UGD/Programacion_Avanzada/refs/heads/main/TP10/datos_meteorologicos.csv')
```

```python
print("Observaciones iniciales del DataFrame:")
print(df.head())
```

```python
print("\nInformación del DataFrame:")
df.info()
```

```python
print("\nNombres de las columnas:")
print(df.columns)
```

```python
fechas = df['Fecha'].to_numpy()
temperaturas = df['Temperatura'].to_numpy(dtype=float)
precipitaciones = df['Precipitación'].to_numpy(dtype=float)
humedad = df['Humedad'].to_numpy(dtype=float)
```

```python
# Calcular el promedio para cada métrica
temp_promedio_imputacion = np.nanmean(temperaturas)
precip_promedio_imputacion = np.nanmean(precipitaciones)
hum_promedio_imputacion = np.nanmean(humedad)

# Reemplazar los NaN con el promedio correspondiente
temperaturas[np.isnan(temperaturas)] = temp_promedio_imputacion
precipitaciones[np.isnan(precipitaciones)] = precip_promedio_imputacion
humedad[np.isnan(humedad)] = hum_promedio_imputacion
```

```python
temperatura_promedio = np.mean(temperaturas)
total_precipitaciones = np.sum(precipitaciones)
maxima_humedad = np.max(humedad)

# Obtener el ÍNDICE del valor máximo/mínimo
indice_max_temp = np.argmax(temperaturas)
indice_min_temp = np.argmin(temperaturas)

# Usar el índice para encontrar la fecha correspondiente
fecha_mas_calurosa = fechas[indice_max_temp]
fecha_mas_fria = fechas[indice_min_temp]
```

```python
# Crear un diccionario con los resultados
resultados = {
    'Metrica': [
        'Temperatura Promedio (°C)',
        'Total de Precipitaciones (mm)',
        'Máxima Humedad Registrada (%)',
        'Fecha más Calurosa',
        'Fecha más Fría'
    ],
    'Valor': [
        f"{temperatura_promedio:.2f}",
        f"{total_precipitaciones:.2f}",
        f"{maxima_humedad:.2f}",
        fecha_mas_calurosa,
        fecha_mas_fria
    ]
}
df_resultados = pd.DataFrame(resultados)

# Exportar a un archivo CSV
df_resultados.to_csv('analisis_meteorologico.csv', index=False)

print("\nResultados:")
print(df_resultados)
```
