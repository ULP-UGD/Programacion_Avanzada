Ejercicio 4

Disponés de un dataset con las siguientes variables:
● Edad: edad de los clientes
● Compra: 1 si realizó una compra, 0 si no
Tareas:
1. Ajustar un modelo de regresión logística para predecir la probabilidad de compra según la edad.
2. Dibujar la curva logística y los datos reales.
3. Calcular la probabilidad de compra para una persona de 30 años.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

import warnings
warnings.filterwarnings("ignore", category=UserWarning)

%matplotlib inline
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/ULP-UGD/Programacion_Avanzada/refs/heads/main/TP12/edad_compra.csv')
X = data[['Edad']]
y = data['Compra']

modelo = LogisticRegression()
modelo.fit(X, y)

X_test = np.linspace(15, 55, 300).reshape(-1, 1)
y_prob = modelo.predict_proba(X_test)[:, 1]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X_test, y_prob, 'r-', label='Curva logística')
plt.title('Probabilidad de compra vs. Edad', fontsize=14)
plt.xlabel('Edad (años)', fontsize=12)
plt.ylabel('Probabilidad de compra', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

prob_30 = modelo.predict_proba(pd.DataFrame([[30]], columns=['Edad']))[0, 1]
print(f"Probabilidad de compra a los 30 años: {prob_30:.2%}")
```
