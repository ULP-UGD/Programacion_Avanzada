Ejercicio 3

El archivo estudiantes.csv contiene:

● Horas_estudio: cantidad de horas semanales dedicadas al estudio

● Nota_final: nota obtenida en el examen final

Tareas:
1. Crear un modelo de regresión lineal que relacione Horas_estudio con Nota_final.
2. Evaluar el coeficiente de determinación R2R^2R2.
3. Determinar si el modelo es adecuado para predecir la nota de un estudiante que estudia 12 horas por semana.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score

%matplotlib inline
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/ULP-UGD/Programacion_Avanzada/refs/heads/main/TP12/estudio_nota.csv')
X = data[['Horas_estudio']]
y = data['Nota_final']

modelo = LinearRegression()
modelo.fit(X, y)

r2 = modelo.score(X, y)
print(f"Coeficiente de determinación R²: {r2:.4f}")
```

```python
horas_12 = pd.DataFrame([[12]], columns=['Horas_estudio'])
nota_predicha = modelo.predict(horas_12)[0]
print(f"Nota predicha para 12 horas de estudio: {nota_predicha:.2f}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, modelo.predict(X), 'r-', label='Regresión lineal')
plt.title('Rendimiento académico vs. Horas de estudio', fontsize=14)
plt.xlabel('Horas de estudio semanales', fontsize=12)
plt.ylabel('Nota final', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
