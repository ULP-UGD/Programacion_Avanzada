Ejercicio 2

Un dataset contiene datos de:
● Edad: en años
● Salario: en pesos
Tareas:
1. Ajustar un modelo de regresión lineal simple para predecir el Salario en función de la Edad.
2. Agregar nuevas edades de prueba (por ejemplo: 25, 35, 45, 55 años) y estimar sus salarios.
3. Mostrar la gráfica con los datos reales, la regresión y las predicciones.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

plt.style.use('ggplot')
%matplotlib inline
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/ULP-UGD/Programacion_Avanzada/refs/heads/main/TP12/edad_salario.csv')
X = data[['Edad']]
y = data['Salario']
```

```python
modelo = LinearRegression()
modelo.fit(X, y)

nuevas_edades = pd.DataFrame([25, 35, 45, 55], columns=['Edad'])
predicciones = modelo.predict(nuevas_edades)

print("Predicciones de salario:")
for edad, salario in zip(nuevas_edades['Edad'], predicciones):
    print(f"Edad {edad} años: ${salario:.2f}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, modelo.predict(X), 'r-', label='Regresión lineal')
plt.scatter(nuevas_edades, predicciones, color='green', s=100, marker='X', label='Predicciones')
plt.title('Relación Edad vs. Salario', fontsize=14)
plt.xlabel('Edad (años)', fontsize=12)
plt.ylabel('Salario ($)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
