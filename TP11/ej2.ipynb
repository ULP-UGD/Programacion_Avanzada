Representa las ventas mensuales de una librería.

```
Meses: ["Ene", "Feb", "Mar", "Abr", "May"]
Ventas: [150, 200, 180, 220, 170]
```

* Etiqueta los ejes
* Asigna un color diferente a cada barra
* Muestra el valor de cada barra sobre ella

````python
import matplotlib.pyplot as plt

Meses = ["Ene", "Feb", "Mar", "Abr", "May"]
Ventas = [150, 200, 180, 220, 170]
````

````python
plt.figure(figsize=(8,4))
bars = plt.bar(Meses, Ventas, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum'])
plt.xlabel("Mes")
plt.ylabel("Ventas")
plt.title("Ventas mensuales de la librería")
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')
plt.show()
````
