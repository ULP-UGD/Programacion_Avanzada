Utiliza uno de los gráficos anteriores y:

* Cambia el estilo de fondo con plt.style.use('ggplot') o similar
* Guarda el gráfico como imagen PNG
* Agrega una cuadrícula al gráfico
* Aumenta el tamaño del texto y de los marcadores

```python
import matplotlib.pyplot as plt

Meses = ["Ene", "Feb", "Mar", "Abr", "May"]
Ventas = [150, 200, 180, 220, 170]


plt.style.use('ggplot')

plt.figure(figsize=(10, 6))
bars = plt.bar(Meses, Ventas, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum'])
plt.xlabel("Mes", fontsize=12)
plt.ylabel("Ventas", fontsize=12)
plt.title("Ventas mensuales de la librería", fontsize=14)


plt.grid(True, axis='y', linestyle='--', alpha=0.7)


for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom', ha='center', fontsize=10)

plt.savefig('ventas_libreria.png')

plt.show()
```
