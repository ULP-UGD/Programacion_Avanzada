```python
df1=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Programacion Avanzada/Dataset/Datos_Ventas_Tienda.csv')
df2=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Programacion Avanzada/Dataset/Datos_Ventas_Tienda2.csv')
```

```python
df2.head()
```

```python
df=pd.concat([df1,df2],ignore_index=True)
df
```

```python
df.info()
```

```python
type(df['Fecha'][0])
```

```python
df['Fecha']=pd.to_datetime(df['Fecha'])
df.info()
```

```python
#¿Cual es el producto mas vendido?
prod_mas_vendido=df.groupby('Producto')['Cantidad'].sum().idxmax()
print(f'El producto mas vendido es: {prod_mas_vendido}')
```

```python
#¿Cuál es el mes con más ventas?
meses=[]
for fecha in df['Fecha']:
  meses.append(fecha.month)
df['Mes']=meses
df
```

```python
ventas_por_mes=df.groupby('Mes')['Total Venta'].sum().idxmax()
print(f'Hubo mas ventas el mes {ventas_por_mes}')
```

```python
#Agrupa los datos por categorias de productos
df.groupby('Producto')['Total Venta'].sum().sort_values(ascending=False)
```
